```python
import numpy as np
import matplotlib.pyplot as plt
array = np.random.rand(1000)
```

```python
plt.plot(array)
```

```python
var = np.std(array)
mean = np.mean(array)
```

```python
def log_likely(x, array):
    var = np.std(array)
    mean = np.mean(array)
    p1 = np.log(1/(var*np.sqrt(2*np.pi)))
    p2 = (x-mean)**2/(2*var**2)
    return mean, var, p1-p2
```

```python
log_likely(0.99, array)
```
