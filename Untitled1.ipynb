```python
import random
import functools
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARMA as smARMA
import statsmodels.api as sm
```

```python
l = 2000
test = np.random.rand(l)
print(test)
```

```python
plt.plot(test)
```

```python
models = {}
for i in range(20):
    try:
        py_arma = smARMA(test, [i,i])
        py_arma.fit().summary()
        models[i]=py_arma.fit()
    except:
        print(str(i)+" failed")
```

```python
hists = {}
for k,v in models.items():
    hists[k] = sm.tsa.arma_generate_sample(ar = np.append(1., -v.arparams),
                                       ma = np.append(1., v.maparams),
                                       nsample = l,
                                       sigma = np.sqrt(v.sigma2),
                                       burnin = 2*k)
```

```python
for k,v in models.items():
    print(v.summary())
```
